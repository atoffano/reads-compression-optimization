```python
import matplotlib.pyplot as plt
import seaborn as sns

from method_2_jules import launch_method_2


FILES = ["data/ecoli_100Kb_reads_5x.fasta",
        "data/ecoli_100Kb_reads_10x.fasta",
        "data/ecoli_100Kb_reads_20x.fasta",
        "data/ecoli_100Kb_reads_40x.fasta",
        "data/ecoli_100Kb_reads_80x.fasta",
        "data/ecoli_100Kb_reads_120x.fasta",]
file_size = [file[23:-6] for file in FILES]
KMER_SIZE = [2,3,4,5,6,7,8,9,10]
```

### Metrics for random method :

#### Load Metrics (random)

```python
rate_by_file = []
ram_by_file = []
time_by_file = []

for file in FILES:
    kwargs = {
    "input":file,
    "output":"data/method_2_out.fasta",
    "delete_output":True,
    "method":"random",
    "size":5,
    }
    log = launch_method_2(**kwargs)
    rate_by_file.append(float(log["rate"]))
    ram_by_file.append(int(log["mem_usage"][0]))
    time_by_file.append(log["exec_time"])

rate_by_size = []
ram_by_size = []
time_by_size = []

for size in KMER_SIZE:
    kwargs = {
    "input":"data/ecoli_100Kb_reads_40x.fasta",
    "output":"data/method_2_out.fasta",
    "delete_output":True,
    "method":"random",
    "size":size,
    }
    log = launch_method_2(**kwargs)
    rate_by_size.append(float(log["rate"]))
    ram_by_size.append(int(log["mem_usage"][0]))
    time_by_size.append(log["exec_time"])
```

#### Vizualize Metrics (random)

```python

plt.figure(figsize=(8,6))
size_comparisons= sns.scatterplot(
    x=file_size, y=rate_by_file,
    legend="brief",
).set(title="Compression rates by file random method size of 5"
      ,xlabel="file size", ylabel="compression rate")

plt.figure(figsize=(8,6))
size_comparisons= sns.scatterplot(
    x=file_size, y=ram_by_file,
    legend="brief",
).set(title="RAM use by file random method size of 5"
      ,xlabel="file size", ylabel="cRAM used (B)")

plt.figure(figsize=(8,6))
size_comparisons= sns.scatterplot(
    x=file_size, y=time_by_file,
    legend="brief",
).set(title="Exec time by file random method size of 5"
      ,xlabel="file size", ylabel="time (s)")


plt.figure(figsize=(8,6))
size_comparisons= sns.scatterplot(
    x=KMER_SIZE, y=rate_by_size,
    legend="brief",
).set(title="Compression rates by kmer size random method 40x file"
      ,xlabel="file size", ylabel="compression rate")

plt.figure(figsize=(8,6))
size_comparisons= sns.scatterplot(
    x=KMER_SIZE, y=ram_by_size,
    legend="brief",
).set(title="RAM use by kmer size random method 40x file"
      ,xlabel="file size", ylabel="cRAM used (B)")

plt.figure(figsize=(8,6))
size_comparisons= sns.scatterplot(
    x=KMER_SIZE, y=time_by_size,
    legend="brief",
).set(title="Exec time by kmer size random method 40x file"
      ,xlabel="file size", ylabel="time (s)")
```
