```python
import matplotlib.pyplot as plt
import seaborn as sns

from kmer_sort import launch_method_2


FILES = ["data/ecoli_100Kb_reads_5x.fasta",
        "data/ecoli_100Kb_reads_10x.fasta",
        "data/ecoli_100Kb_reads_20x.fasta",
        "data/ecoli_100Kb_reads_40x.fasta",
        "data/ecoli_100Kb_reads_80x.fasta",
        "data/ecoli_100Kb_reads_120x.fasta",]
file_size = [file[23:-6] for file in FILES]
KMER_SIZE = [2,3,4,5,6,7,8,9,10]
```

### Method comparison :

#### Load Metrics (random)

```python
random_rate_by_file = []
random_ram_by_file = []
random_time_by_file = []

for file in FILES:
    kwargs = {
    "input":file,
    "output":"data/method_2_out.fasta",
    "delete_output":True,
    "method":"random",
    "size":5,
    }
    log = launch_method_2(**kwargs)
    random_rate_by_file.append(float(log["rate"]))
    random_ram_by_file.append(int(0 if not log["mem_usage"] else log["mem_usage"][0]))
    random_time_by_file.append(log["exec_time"])

random_rate_by_size = []
random_ram_by_size = []
random_time_by_size = []

for size in KMER_SIZE:
    kwargs = {
    "input":"data/ecoli_100Kb_reads_40x.fasta",
    "output":"data/method_2_out.fasta",
    "delete_output":True,
    "method":"random",
    "size":size,
    }
    log = launch_method_2(**kwargs)
    random_rate_by_size.append(float(log["rate"]))
    random_ram_by_size.append(int(0 if not log["mem_usage"] else log["mem_usage"][0]))
    random_time_by_size.append(log["exec_time"])
```

#### Load metrics occurence maxi method

```python
occ_max_rate_by_file = []
occ_max_ram_by_file = []
occ_max_time_by_file = []

for file in FILES:
    kwargs = {
    "input":file,
    "output":"data/method_2_out.fasta",
    "delete_output":True,
    "method":"occurence-max",
    "size":3,
    }
    log = launch_method_2(**kwargs)
    occ_max_rate_by_file.append(float(log["rate"]))
    occ_max_ram_by_file.append(int(0 if not log["mem_usage"] else log["mem_usage"][0]))
    occ_max_time_by_file.append(log["exec_time"])

occ_max_rate_by_size = []
occ_max_ram_by_size = []
occ_max_time_by_size = []

for size in KMER_SIZE:
    kwargs = {
    "input":"data/ecoli_100Kb_reads_40x.fasta",
    "output":"data/method_2_out.fasta",
    "delete_output":True,
    "method":"occurence-max",
    "size":size,
    }
    log = launch_method_2(**kwargs)
    occ_max_rate_by_size.append(float(log["rate"]))
    occ_max_ram_by_size.append(int(0 if not log["mem_usage"] else log["mem_usage"][0]))
    occ_max_time_by_size.append(log["exec_time"])
```

### Plot metrics

#### Plot metrics random method

```python
plt.figure(figsize=(12,6))
plt.ylim(1, 3)
size_comparisons= sns.scatterplot(
    x=file_size, y=random_rate_by_file,
    legend="brief",
).set(title="Compression rates by file random method size of 5"
      ,xlabel="file size", ylabel="compression rate")


plt.figure(figsize=(12,6))
size_comparisons= sns.scatterplot(
    x=file_size, y=random_ram_by_file,
    legend="brief",
).set(title="RAM use by file random method size of 5"
      ,xlabel="file size", ylabel="cRAM used (B)")

plt.figure(figsize=(12,6))
plt.ylim(0.5, 10)
size_comparisons= sns.scatterplot(
    x=file_size, y=random_time_by_file,
    legend="brief",
).set(title="Exec time by file random method size of 5"
      ,xlabel="file size", ylabel="time (s)")

plt.figure(figsize=(12,6))
plt.ylim(1, 3)
size_comparisons= sns.scatterplot(
    x=KMER_SIZE, y=random_rate_by_size,
    legend="brief",
).set(title="Compression rates by kmer size random method 40x file"
      ,xlabel="file size", ylabel="compression rate")

plt.figure(figsize=(12,6))
size_comparisons= sns.scatterplot(
    x=KMER_SIZE, y=random_ram_by_size,
    legend="brief",
).set(title="RAM use by kmer size random method 40x file"
      ,xlabel="file size", ylabel="cRAM used (B)")

plt.figure(figsize=(12,6))
plt.ylim(0.5, 10)
size_comparisons= sns.scatterplot(
    x=KMER_SIZE, y=random_time_by_size,
    legend="brief",
).set(title="Exec time by kmer size random method 40x file"
      ,xlabel="file size", ylabel="time (s)")
```
